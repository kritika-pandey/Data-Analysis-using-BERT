This is data of issues raised by clients of different banks in USA. 

Source : https://www.consumerfinance.gov/data-research/consumer-complaints/#download-the-data > Download all complaint data | CSV 

```python
##importing modules
import pandas as pd
from transformers import pipeline 
nlp = pipeline("sentiment-analysis")
```

```python
pip install tensorflow
```

```python
df = pd.read_csv('complaints.csv')   ##importing data into dataframe
```

```python
df.head()
```

```python
df.shape
```

```python
##Cleaning (NaN)s from the data set
df = df.dropna(how = 'any') 
```

```python
df.shape
```

```python
df.head()
```

```python
df.reset_index(inplace = True)
```

```python
df.head()
```

```python
df.drop(['index'], axis = 1,inplace = True) 
df.head()
```

```python
##extracting sentiment of issues raised (df['Issue']) and saving into a new DataFrame 'temp'
temp = nlp(list(df['Issue']))
```

```python
temp = pd.DataFrame(temp)
```

```python
#observe that temp has two columns 'label' (indicating the sentiment) and 'score'(precision of sentiment analysed)
temp.head()
```

```python
df['Sentiment Issue'] = temp['label']
```

```python
df.head()
```

```python
import datetime
```

```python
df['Date received'] = pd.to_datetime(df['Date received'])
```

Observing the negative sentiments on monthly basis

```python
df[df['Sentiment Issue']=='NEGATIVE'][['Sentiment Issue', 'Date received']].groupby(pd.Grouper(key='Date received', freq="M")).count().plot(kind = 'bar')
```

Observing the positive sentiments on monthly basis

```python
df[df['Sentiment Issue']=='POSITIVE'][['Sentiment Issue', 'Date received']].groupby(pd.Grouper(key='Date received', freq="M")).count().plot(kind = 'bar')
```

Observing cumulative graph of increase in negative sentiments

```python
df[df['Sentiment Issue']=='NEGATIVE'][['Sentiment Issue', 'Date received']].groupby(pd.Grouper(key='Date received', freq="M")).count().cumsum().plot()
```

Observing cumulative graph of increase in positive sentiments

```python
df[df['Sentiment Issue']=='POSITIVE'][['Sentiment Issue', 'Date received']].groupby(pd.Grouper(key='Date received', freq="M")).count().cumsum().plot()
```
